```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn import *
import re
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')
tqdm.pandas()
import tensorflow as tf
import tensorflow.keras as keras
from tensorflow.keras.layers import *
import datetime
```

```python
df = pd.read_csv(r'C:\Users\AA2095\3D Objects\data\Sets\music.csv')
```

```python
df['splitted_Lyrics']=df['Lyrics'].progress_apply(lambda x: re.findall('[A-Z][^A-Z]*',str(x)))
```

```python
band_gb=df.groupby('Artist')['splitted_Lyrics'].sum().reset_index()
```

```python
snoop=band_gb.iloc[5137]['splitted_Lyrics']
```

```python
nirvana_texts=band_gb.query('Artist == "Nirvana"')['splitted_Lyrics'][4100]
```

```python
tokenizer = keras.preprocessing.text.Tokenizer()
```

```python
tokenizer.fit_on_texts(snoop)
```

```python
total_sents=len(tokenizer.word_index)+1
```

```python
input_seqen =[]
```

```python
for line in snoop:

    token_list = tokenizer.texts_to_sequences([line])[0]

    for i in range(1,len(token_list)):

        n_gram_seq = token_list[:i+1]

        input_seqen.append(n_gram_seq)
```

```python
max_len =max([len(x) for x in input_seqen])
```

```python
input_seq_padded = np.array(keras.preprocessing.sequence.pad_sequences(input_seqen,maxlen=max_len,padding='pre'))
```

```python
xs = input_seq_padded[:,:-1]
labels = input_seq_padded[:,-1]
```

```python
ys = keras.utils.to_categorical(labels)
```

```python
model = keras.Sequential([Embedding(total_sents,256,input_length=max_len-1),
(GRU(500)),
Dense(total_sents,activation='softmax')])
```

```python
model.compile(loss='categorical_crossentropy',optiizer='Adam',metrics=['accuracy'])
```

```python
history = model.fit(xs,ys,epochs=26,verbose=1,batch_size=64)
```
